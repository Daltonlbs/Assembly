```python
def Preprocessing(inst):
    inst = re.sub(r"http\S+", "", inst).lower().replace('.','').replace(',','').replace(';','').replace('-','').replace(':','').replace(')','')
    stopwords = set(nltk.corpus.stopwords.words('portuguese'))
    palavras = [i for i in inst.split() if not i in stopwords]
    return (" ".join(palavras))

def prep_tweets(tweet):
    
    tweet = BeautifulSoup(tweet, "html.parser").get_text()
    tweet = re.sub(r"[^a-zA-Zà-úÀ-Ú0-9]", " ", tweet.lower())
    return tweet

def traduz(pt_text):
    tradutokenizer.src_lang = "pt"
    encoded_pt = tradutokenizer(pt_text, return_tensors="pt")
    generated_tokens = tradumodel.generate(**encoded_pt, forced_bos_token_id=tradutokenizer.get_lang_id("en"))
    return tradutokenizer.batch_decode(generated_tokens, skip_special_tokens=True)

```
